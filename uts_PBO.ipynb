```python
# --- KODE INHERITANCE ---
class TiketWisata:
    def __init__(self, id_tiket, destinasi, harga_dasar):
        self.id_tiket = id_tiket
        self.destinasi = destinasi
        self.harga_dasar = harga_dasar

# Varian 1: Tiket Reguler
class TiketReguler(TiketWisata):
    def __init__(self, id_tiket, destinasi, harga_dasar):
        super().__init__(id_tiket, destinasi, harga_dasar)
        self.kategori = "Reguler"
        self.fitur = "Akses Standar"

# Varian 2: Tiket VIP
class TiketVIP(TiketWisata):
    def __init__(self, id_tiket, destinasi, harga_dasar):
        super().__init__(id_tiket, destinasi, harga_dasar)
        self.kategori = "VIP"
        self.fitur = "Akses Jalur Cepat + Makan Siang"

# Varian 3: Tiket Terusan (Multi-Day)
class TiketTerusan(TiketWisata):
    def __init__(self, id_tiket, destinasi, harga_dasar):
        super().__init__(id_tiket, destinasi, harga_dasar)
        self.kategori = "Terusan"
        self.fitur = "Berlaku untuk 3 Hari Kunjungan"

# --- HASIL RUN ---
print("="*40)
print("HASIL IMPLEMENTASI INHERITANCE")
print("="*40)
reg = TiketReguler("R01", "Candi Borobudur", 50000)
vip = TiketVIP("V01", "Taman Safari", 200000)
trs = TiketTerusan("T01", "Gunung Bromo", 500000)

for t in [reg, vip, trs]:
    print(f"ID: {t.id_tiket} | {t.destinasi} ({t.kategori}) -> Fitur: {t.fitur}")
```

```python
# --- KODE POLIMORFISME ---
class MesinHitung:
    def hitung_bayar(self, harga, jumlah):
        return harga * jumlah

# Varian 1: Diskon Reguler (Potongan 10% jika beli > 5 tiket)
class HitungReguler(MesinHitung):
    def hitung_bayar(self, harga, jumlah):
        total = super().hitung_bayar(harga, jumlah)
        return total * 0.9 if jumlah > 5 else total

# Varian 2: Pajak VIP (Tambahan Pajak Layanan 15%)
class HitungVIP(MesinHitung):
    def hitung_bayar(self, harga, jumlah):
        total = super().hitung_bayar(harga, jumlah)
        return total * 1.15

# Varian 3: Biaya Admin Tiket Terusan (Tambah Biaya Cetak Kartu 25.000)
class HitungTerusan(MesinHitung):
    def hitung_bayar(self, harga, jumlah):
        total = super().hitung_bayar(harga, jumlah)
        return total + 25000

# --- HASIL RUN ---
print("\n" + "="*40)
print("HASIL IMPLEMENTASI POLIMORFISME")
print("="*40)
pilihan_hitung = [HitungReguler(), HitungVIP(), HitungTerusan()]
label = ["Tiket Reguler (Beli 10)", "Tiket VIP (Beli 1)", "Tiket Terusan (Beli 1)"]

for i in range(len(pilihan_hitung)):
    qty = 10 if i == 0 else 1
    hasil = pilihan_hitung[i].hitung_bayar(100000, qty)
    print(f"{label[i]:<25} : Rp{hasil:,.0f}")
```

```python
# --- KODE ENKAPSULASI ---
class KeamananData:
    def __init__(self):
        self.__kode_admin = "2026"      # Private (Data Rahasia 1)
        self.__harga_pokok = 45000      # Private (Data Rahasia 2)
        self.__saldo_masuk = 0          # Private (Data Rahasia 3)

    # Getter: Hanya melihat harga pokok
    def cek_harga_pokok(self):
        return self.__harga_pokok

    # Setter 1: Mengubah harga pokok (Harus pakai PIN)
    def update_harga_pokok(self, harga_baru, pin):
        if pin == self.__kode_admin:
            self.__harga_pokok = harga_baru
            print("[Sukses] Harga pokok berhasil diperbarui.")
        else:
            print("[Gagal] Akses Ditolak: PIN Admin Salah!")

    # Setter 2: Mencatat Saldo (Hanya angka positif)
    def setor_saldo(self, jumlah):
        if jumlah > 0:
            self.__saldo_masuk += jumlah

    # Getter: Cek total saldo
    def total_pendapatan(self):
        return self.__saldo_masuk

# --- HASIL RUN ---
print("\n" + "="*40)
print("HASIL IMPLEMENTASI ENKAPSULASI")
print("="*40)
sistem = KeamananData()

print(f"Harga Pokok Saat Ini: Rp{sistem.cek_harga_pokok()}")
# Coba ubah harga dengan PIN salah
sistem.update_harga_pokok(50000, "0000")
# Coba ubah harga dengan PIN benar
sistem.update_harga_pokok(55000, "2026")
print(f"Harga Pokok Baru    : Rp{sistem.cek_harga_pokok()}")
```
